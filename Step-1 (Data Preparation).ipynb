Make Column to aggregrate Sentiment Value

```python
df['power_of_senti'] = df['positive_senti'] + df['negative_senti']
```

```python
df.head()
```

```python
def classify_sentiment(x):
    if (x['power_of_senti'] >= 1 and x['power_of_senti'] < 4):
        return 'Positive'
    elif (x['power_of_senti'] >= 4):
        return 'High_Positive'
    elif (x['power_of_senti'] <= -1 and x['power_of_senti'] > -4):
        return 'Negative'
    elif (x['power_of_senti'] == 0):
        return 'Neutral'
    else:
        return 'Highly Negative'
df = df.assign(avg_senti=df.apply(classify_sentiment, axis=1))
```
